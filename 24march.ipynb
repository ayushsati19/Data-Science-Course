Nonflavanoid phenols: Non-flavanoid phenolic compounds are another type of phenolic compound that can contribute to the taste, aroma, and overall quality of the wine.

Proanthocyanins: Proanthocyanins are a specific type of flavanoid that can contribute to the color, flavor, and overall quality of the wine.

**2.Handling missing Values in the dataset**
This dataset file has no missing values.But if there were some missing values we could use one of the methods:-

Deletion: This method involves removing the entire row or column where the missing value is located. However, this approach can result in a loss of information and reduce the overall size of the dataset.

Imputation: This method involves replacing the missing value with a value that is derived from the other values in the dataset. This can be done using various techniques, such as mean imputation, median imputation, or regression imputation.

Model-based imputation: This method involves using a statistical model to predict the missing values based on the values in the dataset. This approach can provide more accurate imputations but requires more complex methods.

Here are some advantages and disadvantages of the three main methods for handling missing values in a dataset:

Deletion:

Advantages:

Easy to implement.
Can be useful when the missing values are randomly distributed across the dataset.

Disadvantages:

Can lead to loss of information and reduce the size of the dataset.
Can introduce biases and distort the distribution of the data if the missing values are not randomly distributed.

Imputation:

Advantages:

Can provide a complete dataset for analysis.
Can help to preserve the distribution and relationships among the variables.
Can be useful for exploratory data analysis and building predictive models.

Disadvantages:

Can introduce biases if the missing values are not randomly distributed.
May not accurately capture the true value of the missing data.
Can be sensitive to outliers and extreme values in the data.

Model-based imputation:

Advantages:

Can provide more accurate imputations by using a statistical model.
Can help to preserve the distribution and relationships among the variables.
Can be useful for complex datasets with multiple missing values.

Disadvantages:

Can be computationally expensive and require significant resources.
Can be sensitive to the choice of statistical model and assumptions.
May not work well for datasets with highly skewed or non-normal distributions.

#### **Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?**

```python
#performing pca on wine quality data set
data_file.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(data_file)

# calculate the eigenvalues
eigenvalues = pca.explained_variance_

# calculate the total sum of eigenvalues
total_sum = np.sum(eigenvalues)

# calculate the cumulative sum of eigenvalues
cumulative_sum = np.cumsum(eigenvalues)

# calculate the proportion of variance explained
prop_var = cumulative_sum / total_sum

# identify the number of principal components required to explain 90% of the variance
n_components = np.argmax(prop_var >= 0.9) + 1

print("Number of principal components required to explain 90% of the variance:", n_components)
```
